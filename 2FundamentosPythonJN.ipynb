```python
### Desafio Python3
'''
====  REQUISITOS ======
Faça um pequeno programa em Python3, para atender a um cliente de  uma empresa de CryptoAtivos.

Na aplicação, o cliente deve inserir seu nome e telefone de contato.  Após o cadastro, com a inserção dos dados, 
serão gerados 5 numeros das sorte, que devem ser exibidos, de modo ordenado crescente. A senha de acesso será a 
concatenação do menor numero e o maior numero.

Após conclusos, deve ser exibido os dados digitados e uma saudação de “ prezado Sr (a) nome, obrigado pelo seu cadastro, 
os números da sorte o telefone de contato e a senha do cliente.
'''
print("\n--- DESAFIO PYTHON 3 - 3 === \n")
```

```python
# Solução proposta
import random

def main():
    print("=== CADASTRO - CRYPTOATIVOS ===")
    
    # 1. Coleta de dados do cliente
    nome = input("Digite seu nome completo: ").strip()
    telefone = input("Digite seu telefone de contato (com DDD): ").strip()
    
    # 2. Geração de 5 números da sorte (valores entre 1 e 100, sem repetição)
    # Usamos random.sample para garantir que os 5 números sejam distintos
    numeros_da_sorte = random.sample(range(1, 101), 5)
    
    # Ordena os números em ordem crescente
    numeros_da_sorte.sort()
    
    # 3. Definição da senha (concatenação do menor e do maior número)
    # Como a lista já está ordenada, o menor é o primeiro [0] e o maior é o último [-1]
    menor_numero = numeros_da_sorte[0]
    maior_numero = numeros_da_sorte[-1]
    senha_acesso = f"{menor_numero}{maior_numero}"
    
    # Converte os números para uma string bonita para exibição (ex: 12, 25, 47...)
    numeros_formatados = ", ".join(map(str, numeros_da_sorte))
    #Inserir figuras para usabilidade
    # Usando o código Unicode (substitua o U+ por 000)
    print("Acesse com a sua chave: \U0001F511")
    # 4. Exibição do relatório final e saudação
    print("\n" + "="*50)
    print(f"Prezado(a) Sr(a) {nome}, obrigado pelo seu cadastro!")
    print("="*50)
    print(f"📱 Telefone de contato: {telefone}")
    print(f"🍀 Números da sorte (ordenados): {numeros_formatados}")
    print(f"🔑 Senha de acesso gerada: {senha_acesso}")
    print("="*50)

#print( __name__)
if __name__ == "__main__":
    main()
```

```python
### Dicionários 
"""
Um dicionário (dict) é uma coleção de dados mutável e ordenada (a partir do Python 3.7);
Armazena elementos no formato de chave: valor (key: value).
Imagine um dicionário real: você procura por uma palavra (chave) para encontrar o seu significado (valor). 
No Python é a mesma coisa, você usa a chave para acessar o valor rapidamente.
"""
```

```python
# Criando Dicionários
# Os dicionários são definidos utilizando chaves {}
# e os pares são separados por dois pontos :.

# Dicionário vazio
dicionario_vazio = {}

# Dicionário de um produto (Chaves são strings,
# valores são tipos variados)
produto = {
    "nome": "Teclado Mecânico",
    "preco": 299.90,
    "em_estoque": True,
    "cores_disponiveis": ["Preto", "Branco"]
}

# Criando com a função dict()
usuario = dict(nome="Francisco", idade=30, cargo="Prof")
```

```python
#  Acessando Valores
# Para acessar um valor, você usa o nome do dicionário 
# seguido da chave entre colchetes [] ou o método .get().
carro = {
    "marca": "Renault",
    "modelo": "Logan",
    "ano": 2015
}

# Forma direta (Gera erro se a chave não existir)
print(carro["modelo"])  # Saída: Mustang

# Usando o método .get() (Mais seguro: 
#retorna None ou um padrão se a chave não existir)
print(carro.get("ano"))      # Saída: 2023
print(carro.get("cor", "Não informada"))  
# Saída: Não informada
```

```python
# Adicionando e Modificando Elementos
# Como os dicionários são mutáveis, você pode alterar 
# valores existentes ou adicionar novas chaves diretamente.
aluno = {
    "nome": "Pedro",
    "curso": "Engenharia de Dados"
}

# Modificando um valor existente
aluno["curso"] = "Ciência da Computação"

# Adicionando uma nova chave-valor
aluno["nota_final"] = 9.5

print(aluno)
# Saída: {'nome': 'Pedro', 'curso': 'Ciência da Computação', 'nota_final': 9.5}
```

```python
# Removendo Elementos
# Existem algumas formas de deletar dados de um dicionário:
smartphone = {
    "marca": "Apple",
    "modelo": "iPhone 15",
    "cor": "Azul",
    "bateria": "100%"
}

# 1. Método .pop(): remove a chave e retorna o valor dela
cor_removida = smartphone.pop("cor") 

# 2. Palavra-chave 'del': remove direto e não retorna nada
del smartphone["bateria"]

# 3. Método .popitem(): remove o último elemento inserido
smartphone.popitem()
```

```python
# Iterando (Navegando) em Dicionários
# Você pode percorrer apenas as chaves, 
# apenas os valores ou ambos ao mesmo tempo 
# usando loops for.
capital_estados = {
    "SP": "São Paulo",
    "RJ": "Rio de Janeiro",
    "MG": "Belo Horizonte"
}

# Iterando pelas Chaves (Padrão)
for sigla in capital_estados:
    print(sigla) # Saída: SP, RJ, MG

# Iterando pelos Valores (.values())
for cidade in capital_estados.values():
    print(cidade) # Saída: São Paulo, Rio de Janeiro...

# Iterando por ambos (.items()) - Muito utilizado!
for sigla, cidade in capital_estados.items():
    print(f"A capital de {sigla} é {cidade}.")
```

```python
# Métodos Úteis - trunfos úteis - pcp nas fases de verificação e validação (testes de softwares) - Guardar na Manga:

    # dicionario.keys(): Retorna uma lista com todas as chaves.

    # dicionario.values(): Retorna uma lista com todos os valores.

    #dicionario.clear(): Limpa o dicionário inteiro.

    # dicionario.update(outro_dict): Atualiza o dicionário com pares de outro dicionário (funde os dois).
```

```python
#============ EXTRAS --> integração tuplas (imutável) + dicionários (mutável)
# 1. Usando TUPLAS para chaves e dados que NÃO mudam (Segurança e Performance)
# Uma tupla como chave do dicionário (ID do produto, Categoria)
# Tuplas também controlam o tamanho da caixa do produto (Largura, Altura, Profundidade)
estoque = {
    (101, "Consoles"): {
        "nome": "PlayStation 5",
        "preco": 4500.00,
        "dimensoes": (39, 10, 26) 
    },
    (102, "Jogos"): {
        "nome": "Elden Ring",
        "preco": 299.90,
        "dimensoes": (17, 13, 1)
    }
}

# 2. Usando o .get() para evitar erros se o produto não existir
# Procurando um ID que não existe no estoque
produto_fantasma = estoque.get((999, "Acessórios"), "Produto não encontrado!")
print(f"Busca segura: {produto_fantasma}") 
# Saída: Busca segura: Produto não encontrado!


# 3. Desempacotamento (Unpacking) ao navegar pelo dicionário com .items()
print("\n--- Relatório de Envios (Usando Desempacotamento) ---")
for (id_prod, categoria), dados in estoque.items():
    # Desempacotando a tupla de dimensões que estava guardada no dicionário
    largura, altura, prof = dados["dimensoes"]
    
    print(f"Produto: {dados['nome']} | Categoria: {categoria}")
    print(f"Dimensões para frete: {largura}cm x {altura}cm x {prof}cm")


# 4. Atualizando o dicionário em lote com o método .update()
# Chegou um lote novo com atualização de preço e um novo produto
novos_dados = {
    (101, "Consoles"): {
        "nome": "PlayStation 5 Slim",  # Atualizou o nome
        "preco": 4200.00,              # Atualizou o preço
        "dimensoes": (35, 9, 21)       # Atualizou as dimensões
    },
    (103, "Acessórios"): {             # Novo produto adicionado
        "nome": "Controle DualSense",
        "preco": 450.00,
        "dimensoes": (16, 11, 6)
    }
}

# Mesclando os dados novos no nosso estoque original
estoque.update(novos_dados)

print("\n--- Estoque Atualizado (.update()) ---")
print(f"Total de produtos diferentes no estoque: {len(estoque)}")
print(f"Novo preço do PS5: R$ {estoque[(101, 'Consoles')]['preco']}")
```

```python
# O que observar no exemplo acima? Ele nos mostra na prática:
    # A tupla (101, "Consoles") virou uma chave única. Como listas são mutáveis, o Python proibiria você de usá-las ali. A tupla garante que o ID e a categoria andem sempre juntos e intactos.
    # O .update() poupou linhas de código. Em vez de alterar item por item, ele substituiu o PS5 antigo pelo Slim e já injetou o Controle DualSense automaticamente.
    # O código ficou limpo com o desempacotamento: largura, altura, prof = dados["dimensoes"] evita que você tenha que usar índices confusos como dados["dimensoes"][0].
```

```python
# Desafio Python3 - Extra - Apicação de mundo real - Banco Python3
"""
Para consolidar tudo o que vimos, vamos criar um pequeno Simulador de Caixa Eletrônico. Esse programa utiliza variáveis para o saldo, operadores para cálculos, condicionais para validar saques e um loop para manter o programa rodando até que o usuário decida sair.
Aqui está o código comentado para o discente estudar a estrutura.

Dados do arquivo texto
Significado:
• horas_estudo → horas estudadas pelo aluno - Francisco
• frequencia → frequência nas aulas
• aprovado → 1 = aprovado / 0 = reprovado

"""

```

```python
### # --- Inicialização ---
saldo = 500.0  # Tipo float (dinheiro)
nome_usuario = "Francisco- VIP"  # Tipo string
operacao = ""  # Variável para controlar o loop

print(f"--- Bem-vindo ao Python3 Bank, {nome_usuario}! ---")

# --- Estrutura de Repetição (Loop Principal) ---
while operacao != "3":
    print("\nEscolha uma opção:")
    print("1 - Ver Saldo")
    print("2 - Sacar Dinheiro")
    print("3 - Sair")
    
    operacao = input("\nDigite o número da operação: ")

    # --- Estrutura Condicional ---
    if operacao == "1":
        print(f"Seu saldo atual é: R$ {saldo:.2f}")
    
    elif operacao == "2":
        valor_saque = float(input("Quanto deseja sacar? "))
        
        # Operador de Comparação e Operador Aritmético
        if valor_saque <= saldo:
            saldo -= valor_saque  # Equivalente a: saldo = saldo - valor_saque
            print(f"Saque de R$ {valor_saque:.2f} realizado com sucesso!")
        else:
            print("Saldo insuficiente para esta operação.")
            
    elif operacao == "3":
        print("Encerrando sistema... Obrigado!")
    
    else:
        print("Opção inválida. Tente novamente.")

print("--- Sessão Finalizada ---")
```

```python
# Código Python Completo (IA + ML + Ciência de Dados)
# ---------------------------------------
# Exemplo simples de Data Science + ML
# ---------------------------------------
# Importação das bibliotecas

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# ---------------------------------------
# 1. Ler arquivo CSV
# ---------------------------------------
dados = pd.read_csv("alunos.csv")
# Mostrar os dados
print("Dados carregados:")
print(dados)

# ---------------------------------------
# 2. Separar entradas e saída
# ---------------------------------------
# Variáveis de entrada
X = dados[["horas_estudo", "frequencia"]]
# Variável alvo
y = dados["aprovado"]

# ---------------------------------------
# 3. Dividir dados de treino e teste
# ---------------------------------------
X_treino, X_teste, y_treino, y_teste = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42
)

# ---------------------------------------
# 4. Criar modelo de Machine Learning
# ---------------------------------------
modelo = DecisionTreeClassifier()
# Treinar modelo
modelo.fit(X_treino, y_treino)

# ---------------------------------------
# 5. Fazer previsões
# ---------------------------------------
previsoes = modelo.predict(X_teste)

# ---------------------------------------
# 6. Avaliar desempenho
# ---------------------------------------
acuracia = accuracy_score(y_teste, previsoes)
print("\nPrevisões:")
print(previsoes)
print("\nAcurácia do modelo:")
print(acuracia)

# ---------------------------------------
# 7. Teste com novo aluno (CORRIGIDO)
# ---------------------------------------
# Não usar uma lista pura. Criar um DataFrame com os nomes exatos das colunas
novo_aluno = pd.DataFrame([[4, 70]], columns=["horas_estudo", "frequencia"])
# Assim, garantir que o scikit-learn saiba exatamente que o número 4 corresponde a "horas_estudo"; e
# o 70 corresponde a "frequencia", eliminando o avisos de segurança (UserWarning)

# Agora o predict roda sem gerar nenhum UserWarning
resultado = modelo.predict(novo_aluno)

print("\nNovo aluno:")
if resultado[0] == 1:
    print("Aluno provavelmente APROVADO")
else:
    print("Aluno provavelmente REPROVADO")
```
